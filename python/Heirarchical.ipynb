```python
# import networkx as nx
import numpy as np
arr = np.loadtxt('../true_distance')
arr2 = np.loadtxt('../answedel')
arr2.sort(axis=1)
```

```python
finalanswer = []
n = 100*2
M = np.zeros((n,n))+np.inf
for a in arr:
    i = int(a[0])
    j = int(a[1])
    val = a[2]
    M[i][j] = val
    M[j][i] = val
for i in range(n):
    M[i][i] = np.inf

gp = dict()
for i in range(int(n/2)):
    gp[i] = np.array([i])

index = 100
done = np.array([])
for i in range(100):
    zen = 0
    sorted_array = np.dstack(np.unravel_index(np.argsort(M.ravel()), (n, n)))[0]
    while(True):
        a  = sorted_array[zen]
        zen += 1
        point1, point2 = a
        if np.where(done==point1)[0].shape[0] == 0 and np.where(done==point2)[0].shape[0] == 0:
            break
        else:
            pass
#             print (done, point1, point2)
    print (a)
    finalanswer += [[index, a[0], a[1]]]
    done = np.concatenate([done, sorted(a)])
    gp[index] = np.concatenate([gp[point1], gp[point2]])
    
    for i in range(index):
        sum=0
        terms = 0
        if np.where(done==i)[0].shape[0] > 0 :
            continue
        for p1 in gp[index]:
            for p2 in gp[i]:
                if i==index:
                    continue
                terms += 1
                if p1 > n/2 or p2 > n/2:
                    print ("Cry")
                if M[p1][p2] > 1000000:
                    print (p1, p2,gp[index], gp[i], index, done, i)
                sum += M[p1][p2]
        M[index][i] = sum/terms
        M[i][index] = sum/terms
        M[index][index] = np.inf
    index +=1
```

```python
for i in range(97):
    if not (np.array_equal(finalanswer[i], arr2[i])):
        print (finalanswer[i], arr2[i])
```
